```python
#create a function to clean the webiste
def website_cleaner (website):
    """
    This function cleans up the raw html 
    Input: a URL to investigate
    Output: cleaned text 
    """
    
    #connection error handling 
    try:
        #get the html from the site
        soup_page = BeautifulSoup(urlopen(website).read().decode('utf-8'), 'html.parser')
    except:
        return 
    
    # Remove these two elements from the BS4 object
    for script in soup_page(["script", "style"]):
        script.extract()  
    
    #get text
    text = soup_page.get_text()
    #break into lines
    lines = (line.strip() for line in text.splitlines())
    
    chunks = (phrase.strip() for line in lines for phrase in line.split("  ")) 
    
    def chunk_space(chunk):
        chunk_out = chunk + ' ' 
        return chunk_out  
        
    
    text = ''.join(chunk_space(chunk) for chunk in chunks if chunk).encode('utf-8') 
    text_ori = text.decode("utf-8")
    text = re.sub("[^a-zA-Z.+3]"," ", text_ori)
    text = text.lower().split()  
    stop_words = set(stopwords.words("english"))
    
    text = [w for w in text if not w in stop_words]
    text = list(set(text))
    
    return text
```
