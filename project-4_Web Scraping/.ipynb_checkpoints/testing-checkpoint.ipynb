```python
count=0
for page_number in range(2):
    url_each_page = 'https://www.indeed.co.uk/jobs?q=data+scientist&limit=100&start={}'.format((str(100*page_number)))
    soup_each_page=BeautifulSoup(urlopen(url_each_page).read(),'html.parser')
    
    results =soup_each_page.find_all('div', attrs={'data-tn-component': 'organicJob'})
    for x in results:
       
        job_id = x.find('h2', attrs={"class": "jobtitle"})['id']
        job_title = x.find('a', attrs={'data-tn-element': "jobTitle"}).text.strip().capitalize()
        job_link = "https://www.indeed.sg" + x.find('h2', attrs={"class": "jobtitle"}).find('a')['href']
        result_list.append([job_id, job_title, job_link]) 
        count+=1
print (count)
print (result_list)
```
